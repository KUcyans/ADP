````python
import dfitspy
from matplotlib import pyplot as plt
import numpy as np
from astropy.io import fits
from astropy.nddata import CCDData
from astropy.stats import mad_std
import astropy.stats
import pynot

import ccdproc as ccdp
from pathlib import Path
from ccdproc import ImageFileCollection
import os
import shutil

from photutils.centroids import centroid_quadratic
from photutils.profiles import RadialProfile

# import pynot.calibs
from pynot.calibs import combine_bias_frames

dirEx4 = '/Users/yhjo/Desktop/AstronomicalDataProcessing/Exercise/Ex4/'
dirData = '/Users/yhjo/Desktop/AstronomicalDataProcessing/Exercise/Ex4/DATA'
dirBias = '/Users/yhjo/Desktop/AstronomicalDataProcessing/Exercise/Ex4/DATA/BIAS'
os.listdir(dirData)
os.listdir(dirBias)
````

````python
def combineBias():
    calibrated_path = Path('/Users/yhjo/Desktop/AstronomicalDataProcessing/Exercise/Ex4/DATA/BIAS/')
    previousDir = os.getcwd()
    os.chdir(dirBias)
    bias_images = [m for m in os.listdir(dirBias) if m.endswith('.fits')]
    combined_bias = ccdp.combine(bias_images, unit = 'electron',
                                 method='average',
                                 sigma_clip=True, sigma_clip_low_thresh=2, sigma_clip_high_thresh=5,
                                 sigma_clip_func=np.ma.median, sigma_clip_dev_func=mad_std,
                                 mem_limit=350e6
                                )
    
    combined_bias.meta['combined'] = True
    combined_bias.write(calibrated_path / 'combined_bias.fit', overwrite=True)
    with open('biaslist') as file:
        lines = file.read().splitlines()
        lines = [line.rstrip() for line in lines]
    biascombine = pynot.calibs.combine_bias_frames(lines, output='MASTER_BIAS.fits', overwrite=True)
    os.chdir(previousDir)

combineBias()
````
